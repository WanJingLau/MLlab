```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("DataSetLab2.csv", header=None)
data.head()
```

```python
#extract the columns 
x_train = data[0].values  
y_train = data[1].values

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
#set title
plt.title("Housing Prices")
#set y-axis label
plt.ylabel('Price (in 1000s of dollars)')
#set x-axis label
plt.xlabel('Size (1000sqft)')
plt.show()
```

```python
w = 100
b = 50
print(f"w:{w}")
print(f"b:{b}")
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0] #shape give us how many x we have, can also use len
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b,)
# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
# Set the title
plt.title("Food Truck")
# Set the y-axis label
plt.ylabel('Population of a city')
# Set the x-axis label
plt.xlabel('Profit')
plt.legend()
plt.show() #result doesnot fit the data
```
